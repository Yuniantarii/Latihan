# Contoh Merubah Data Numerical ke Categorical dengan Dictionary

```python
# Membuat kolom baru pada data
data_bunga['data_angka'] = \
data_bunga['kode_bunga'].map({0:'setosa',1:'versicolor',2:'virginica'})
print(data_bunga)
```

```python
data_bunga.tail()
```

```python
data_bunga['nama_bunga']==data_bunga['data_angka']
```

# Merubah data Categorical ke Numerical

```python
data_bunga['jadi_numerical'] = \
data_bunga['data_angka'].map({'setosa':0,'versicolor':1,'virginica':2})
```

```python
data_bunga.head()
```

```python
data_bunga['kode_bunga']==data_bunga['jadi_numerical']
```
